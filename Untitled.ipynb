```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_classification
```

### Reading in the data

```python
df_cat = pd.read_csv('files_for_lab/categorical.csv')
df_cat.columns = df_cat.columns.str.lower()
df_cat
```

```python
df_num = pd.read_csv('files_for_lab/numerical.csv')
df_num.columns = df_num.columns.str.lower()
df_num
```

```python
df_target = pd.read_csv('files_for_lab/target.csv')
df_target.columns = df_target.columns.str.lower()
df_target
```

### firstly encoding the Categorical columns.

```python
df_cat = pd.get_dummies(df_cat)
```

```python
# Concat the data together
df = pd.concat([df_num, df_cat], axis = 1)
df
```

```python
X = df
y = df_target['target_b']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.20, random_state = 42)

smote = SMOTE(random_state = 42)
X_train_sm, y_train_sm = smote.fit_resample(X_train, y_train)

model = RandomForestRegressor(criterion='friedman_mse', max_depth=8, random_state=42, n_jobs=-1)
model.fit(X_train_sm, y_train_sm)
model.score(X_test, y_test)
```
